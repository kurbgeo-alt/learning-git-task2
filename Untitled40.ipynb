```python
# 1. Wczytanie danych głównych
# ==================================================

df = pd.read_csv('fatal-police-shootings-data (5).csv')

df['signs_of_mental_illness'] = df['signs_of_mental_illness'].astype(str).str.lower() == "true"
```

```python
# 2. Analiza ras + choroby psychiczne
# ==================================================

group = df.groupby(['race', 'signs_of_mental_illness']).size().unstack(fill_value=0)
group['total'] = group.sum(axis=1)
group['percent_with_mental_illness'] = group[True] / group['total'] * 100

print(group)

top_race = group['percent_with_mental_illness'].idxmax()
top_value = group['percent_with_mental_illness'].max()

print("\nRasa z najwyższym odsetkiem przypadków z chorobami psychicznymi:")
print(f"{top_race}: {top_value:.2f}%")
```

```python
# 3. Wystąpienia wg dnia tygodnia
# ==================================================

df['date'] = pd.to_datetime(df['date'])
df['day_of_week'] = df['date'].dt.day_name()

ordered_days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
counts_by_day = df['day_of_week'].value_counts().reindex(ordered_days)

plt.figure(figsize=(10,6))
counts_by_day.plot(kind='bar', color='steelblue')
plt.title("Liczba śmiertelnych interwencji policji wg dnia tygodnia")
plt.xlabel("Dzień tygodnia")
plt.ylabel("Liczba interwencji")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

```python
# 4. Pobieranie populacji stanów z Wikipedii
# ==================================================

url_pop = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population"
headers = {"User-Agent": "Mozilla/5.0"}

html = requests.get(url_pop, headers=headers).text
pop_tables = pd.read_html(html)

pop = pop_tables[0].copy()
```

```python
# Flatten MultiIndex
pop.columns = ["_".join([str(x) for x in col if x]) for col in pop.columns]
```

```python
# znajdź kolumnę nazwy stanu
state_col = [c for c in pop.columns if "state" in c.lower() or "name" in c.lower()][0]
```

```python
# znajdź kolumnę populacji
pop_col = [c for c in pop.columns if "pop" in c.lower()][0]
```

```python
# wyczyść populację
pop["population"] = (
    pop[pop_col]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.extract(r"(\d+)")
    .astype(float)
)
```

```python
# normalizacja nazw stanów
pop["state"] = (
    pop[state_col]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.strip()
)
```

```python
# usuń terytoria (nie są w zbiorze)
pop = pop[~pop["state"].str.contains("District|Puerto|Guam|Samoa|Virgin|Mariana", case=False)]
```

```python
# 5. Pobieranie skrótów stanów z Wikipedii
# ==================================================
state_abbrev = {
    "Alabama": "AL", "Alaska": "AK", "Arizona": "AZ", "Arkansas": "AR",
    "California": "CA", "Colorado": "CO", "Connecticut": "CT", "Delaware": "DE",
    "Florida": "FL", "Georgia": "GA", "Hawaii": "HI", "Idaho": "ID",
    "Illinois": "IL", "Indiana": "IN", "Iowa": "IA", "Kansas": "KS",
    "Kentucky": "KY", "Louisiana": "LA", "Maine": "ME", "Maryland": "MD",
    "Massachusetts": "MA", "Michigan": "MI", "Minnesota": "MN",
    "Mississippi": "MS", "Missouri": "MO", "Montana": "MT", "Nebraska": "NE",
    "Nevada": "NV", "New Hampshire": "NH", "New Jersey": "NJ",
    "New Mexico": "NM", "New York": "NY", "North Carolina": "NC",
    "North Dakota": "ND", "Ohio": "OH", "Oklahoma": "OK", "Oregon": "OR",
    "Pennsylvania": "PA", "Rhode Island": "RI", "South Carolina": "SC",
    "South Dakota": "SD", "Tennessee": "TN", "Texas": "TX", "Utah": "UT",
    "Vermont": "VT", "Virginia": "VA", "Washington": "WA",
    "West Virginia": "WV", "Wisconsin": "WI", "Wyoming": "WY"
}

abbr = pd.DataFrame(
    list(state_abbrev.items()),
    columns=["state", "abbreviation"]
)
```

```python
# 6. Łączenie populacji + skrótów stanów
# ==================================================

merged_pop = pop.merge(abbr, on="state", how="left")
```

```python
# 7. Liczenie incydentów na stan
# ==================================================

counts_by_state = df["state"].value_counts().reset_index()
counts_by_state.columns = ["abbreviation", "incidents"]

final = counts_by_state.merge(merged_pop, on="abbreviation", how="left")

missing = final[final["population"].isna()]
if not missing.empty:
    print("\n⚠ Brak dopasowanej populacji dla:")
    print(missing[["abbreviation", "incidents"]])

final["incidents_per_1000"] = final["incidents"] / final["population"] * 1000

print("\nPrzykładowe dane:")
print(final.head())
```
