```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("HRDataset (2).csv", engine="python")

print(data.columns.tolist())

# =========================
# Konwersja dat
# =========================

data['DOB'] = pd.to_datetime(data['DOB'], errors='coerce')
data['DateofHire'] = pd.to_datetime(data['DateofHire'], errors='coerce')
data['DateofTermination'] = pd.to_datetime(data['DateofTermination'], errors='coerce')

# =========================
# Obliczenie wieku
# =========================

today = pd.to_datetime("2019-12-31")
data['Age'] = (today - data['DOB']).dt.days / 365

# =========================
# Upewnienie się, że projekty są liczbą
# =========================

data['SpecialProjectsCount'] = pd.to_numeric(data['SpecialProjectsCount'], errors='coerce')

# =========================
# Korelacja
# =========================

corr = data['Age'].corr(data['SpecialProjectsCount'])
print("Korelacja (r) między wiekiem a liczbą projektów:", corr)

plt.figure()
plt.scatter(data['Age'], data['SpecialProjectsCount'])
plt.title(f"Zależność między wiekiem a liczbą projektów specjalnych (r = {corr:.2f})")
plt.xlabel("Wiek")
plt.ylabel("Liczba projektów specjalnych")
plt.tight_layout()
plt.show()

# Staż pracy
data['EndDate'] = data['DateofTermination'].fillna(today)
data['Tenure'] = (data['EndDate'] - data['DateofHire']).dt.days / 365

# =========================
# 4. Manager vs PerformanceScore
# =========================

manager_perf = data.groupby('ManagerName')['EngagementSurvey'].mean()

plt.figure()
manager_perf.plot(kind='bar')
plt.title("Średnia ocena (EngagementSurvey) wg Managera")
plt.xlabel("Manager")
plt.ylabel("Średnia EngagementSurvey")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# =========================
# 5. RecruitmentSource vs Tenure
# =========================

recruit_tenure = data.groupby('RecruitmentSource')['Tenure'].mean().sort_values(ascending=False)

plt.figure()
recruit_tenure.plot(kind='bar')
plt.title("Średni staż pracy wg źródła rekrutacji")
plt.xlabel("Recruitment Source")
plt.ylabel("Średni staż (lata)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# =========================
# 6. MaritalDesc vs EmpSatisfaction
# =========================

marital_sat = data.groupby('MaritalDesc')['EmpSatisfaction'].mean()

plt.figure()
marital_sat.plot(kind='bar')
plt.title("Średnia satysfakcja wg stanu cywilnego")
plt.xlabel("Stan cywilny")
plt.ylabel("Średnia EmpSatisfaction")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# =========================
# 7. Struktura wieku aktywnych pracowników
# =========================

active = data[data['EmploymentStatus'] == 'Active']

plt.figure()
plt.hist(active['Age'], bins=8)
plt.title("Struktura wieku aktywnych pracowników")
plt.xlabel("Wiek")
plt.ylabel("Liczba pracowników")
plt.tight_layout()
plt.show()

# =========================
# 8. Wiek vs SpecialProjectsCount
# =========================


data['SpecialProjectsCount'] = pd.to_numeric(data['SpecialProjectsCount'], errors='coerce')
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')

corr = data['Age'].corr(data['SpecialProjectsCount'])
print("Korelacja (r) między wiekiem a liczbą projektów:", corr)

plt.figure()
plt.scatter(data['Age'], data['SpecialProjectsCount'])
plt.title(f"Zależność między wiekiem a liczbą projektów specjalnych (r = {corr:.2f})")
plt.xlabel("Wiek")
plt.ylabel("Liczba projektów specjalnych")
plt.tight_layout()
plt.show()
```
