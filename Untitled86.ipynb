```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
# 1. Wczytanie danych
# =========================
X_train = pd.read_csv('samsung_train.txt', sep=r'\s+', header=None)
y_train = pd.read_csv('samsung_train_labels.txt', sep=r'\s+', header=None)

X_test = pd.read_csv('samsung_test.txt', sep=r'\s+', header=None)
y_test = pd.read_csv('samsung_test_labels.txt', sep=r'\s+', header=None)

# =========================
# 2. Skalowanie danych
# =========================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# =========================
# 3. KMeans  metoda okcia
# =========================
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_train_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, marker='o')
plt.xlabel('Liczba klastrw')
plt.ylabel('Inertia')
plt.title('Metoda okcia  KMeans')
plt.show()

# =========================
# 4. KMeans  klasteryzacja
# =========================
kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)
clusters_kmeans = kmeans.fit_predict(X_train_scaled)

# =========================
# 5. PCA  wizualizacja
# =========================
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train_scaled)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters_kmeans, cmap='viridis', s=5)
plt.title('KMeans + PCA')
plt.show()

# =========================
# 6. DBSCAN
# =========================
dbscan = DBSCAN(eps=2.5, min_samples=10)
clusters_dbscan = dbscan.fit_predict(X_train_scaled)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters_dbscan, cmap='plasma', s=5)
plt.title('DBSCAN + PCA')
plt.show()# =========================
# 7. Random Forest
# =========================
rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42, n_jobs=-1
)

rf.fit(X_train_scaled, y_train.values.ravel())
y_pred_rf = rf.predict(X_test_scaled)

print("Random Forest:")
print(classification_report(y_test, y_pred_rf))

# =========================
# 8. SVM
# =========================
svm = SVC(kernel='rbf', C=10, gamma='scale')
svm.fit(X_train_scaled, y_train.values.ravel())
y_pred_svm = svm.predict(X_test_scaled)

print("SVM:")
print(classification_report(y_test, y_pred_svm))
```
