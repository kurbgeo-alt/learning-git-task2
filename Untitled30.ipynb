```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# 1. Wczytanie danych

movies = pd.read_csv("tmdb_movies.csv")
```

```python
# Usuwamy pustą kolumnę indeksu jeśli istnieje
movies = movies.drop(columns=["Unnamed: 0"], errors="ignore")
```

```python
# Wczytanie tabeli gatunków z poprawnymi nagłówkami
genres = pd.read_csv("tmdb_genres.csv", header=None, names=["genre_id", "genre_name"])
```

```python
# 2. Czyszczenie danych gatunków
genres = genres.replace("N/A", pd.NA).dropna()
genres["genre_id"] = pd.to_numeric(genres["genre_id"])
```

```python
# 3. Połączenie tabel

merged = movies.merge(genres, on="genre_id", how="left")
```

```python
# Konwersja na liczby
genres["genre_id"] = pd.to_numeric(genres["genre_id"])
```

```python
# 3. Połączenie tabel

merged = movies.merge(genres, on="genre_id", how="left")
```

```python
#Lista 10 najwyżej ocenianych filmów powyżej 3.kwartyla liczby głosów
vote_count_3rd_quartile = merged["vote_count"].quantile(0.75)

filtered_movies = merged[merged["vote_count"] > vote_count_3rd_quartile]

top_10_movies = filtered_movies.sort_values(
    by="vote_average", ascending=False
).head(10)

print(top_10_movies[["title", "vote_average", "vote_count", "genre_name"]])
```

```python
# Ustalanie kolumny roku
if "release_year" not in merged.columns:

    if "release_date" in merged.columns:
        # Konwersja release_date do datetime
        merged["release_date"] = pd.to_datetime(merged["release_date"], errors="coerce")

        # Wyodrębnienie roku do release_year
        merged["release_year"] = merged["release_date"].dt.year
    else:
        raise ValueError("Brak kolumn release_year i release_date – nie mogę ustalić roku.")

# Usunięcie wierszy bez roku
merged = merged.dropna(subset=["release_year"]).copy()

# Konwersja roku na int
merged["release_year"] = merged["release_year"].astype(int)

# Filtr 2010–2016
movies_2010_2016 = merged[
    (merged["release_year"] >= 2010) &
    (merged["release_year"] <= 2016)
]
# Konwersja daty
merged["release_date"] = pd.to_datetime(merged["release_date"], errors="coerce")

# Wyciąganie roku
merged["release_year"] = merged["release_date"].dt.year

# Usuwamy wiersze bez roku
merged = merged.dropna(subset=["release_year"]).copy()

# 
merged["release_year"] = merged["release_year"].astype(int)
# ===== 4. Grupowanie =====
grouped = (
    movies_2010_2016
    .groupby("release_date")[["revenue", "budget"]]
    .mean()
    .reset_index()
)

print(grouped)   
print(movies_2010_2016.head())

# Wykres
 #Czyszczenie stanu matplotlib
plt.clf()
plt.cla()
plt.close('all')

fig, ax1 = plt.subplots(figsize=(12, 6))

# Kolumny – średnie przychody
sns.barplot(
    data=grouped,
    x="release_date",
    y="revenue",
    ax=ax1,
    color="skyblue",
    label="Średni przychód"
)

# Linia – średni budżet
ax2 = ax1.twinx()
sns.lineplot(
    data=grouped,
    x="release_date",
    y="budget",
    ax=ax2,
    color="red",
    linewidth=2,
    label="Średni budżet"
)

ax1.set_xlabel("Rok")
ax1.set_ylabel("Średni przychód")
ax2.set_ylabel("Średni budżet")

# Legenda POZA obszarem osi
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
plt.legend(lines_1 + lines_2, labels_1 + labels_2,
           loc="upper right", bbox_to_anchor=(1.25, 1))

plt.title("Średni przychód i budżet filmów (2010–2016)")
plt.tight_layout()
plt.show()















    
    
    









```

```python
genre_counts = merged["genre_name"].value_counts()
most_common_genre = genre_counts.idxmax()
count_most_common = genre_counts.max()

print("Najczęstszy gatunek:", most_common_genre)
print("Liczba filmów:", count_most_common)
```

```python
#Film,którego gatunek
genre_runtime = merged.dropna(subset=["runtime", "genre_name"])
genre_longest = (
    genre_runtime.groupby("genre_name")["runtime"]
    .mean()
    .sort_values(ascending=False)
)

longest_genre = genre_longest.index[0]
print("Najdłuższy średni runtime ma gatunek:", longest_genre)
```

```python
#Histogram czasu trwania filmów
plt.figure(figsize=(10, 6))
sns.histplot(
    genre_runtime[genre_runtime["genre_name"] == longest_genre]["runtime"],
    bins=30,
    kde=True,
    color="purple"
)
plt.title(f"Histogram czasu trwania filmów – {longest_genre}")
plt.xlabel("Czas trwania (min)")
plt.tight_layout()
plt.show()
```
