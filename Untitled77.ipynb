```python
# NOWE MODELE – WIĘCEJ WARSTW SPLOTOWYCH + REGULARYZACJA
# =========================

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Dense, Dropout,
    BatchNormalization, GlobalAveragePooling2D
)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import RMSprop
```

```python
# MODEL 5 – głębsza CNN + BatchNorm + L2 + Dropout
# =========================

model_5 = Sequential()

model_5.add(Conv2D(32, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4),
                   input_shape=(150,150,3)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(Conv2D(64, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(Conv2D(128, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(Conv2D(256, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(GlobalAveragePooling2D())
model_5.add(Dense(256, activation='relu'))
model_5.add(Dropout(0.5))
model_5.add(Dense(1, activation='sigmoid'))

model_5.compile(
    loss='binary_crossentropy',
    optimizer=RMSprop(learning_rate=1e-4),
    metrics=['accuracy']
)
# MODEL 5 – głębsza CNN + BatchNorm + L2 + Dropout
# =========================

model_5 = Sequential()

model_5.add(Conv2D(32, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4),
                   input_shape=(150,150,3)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(Conv2D(64, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(Conv2D(128, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(Conv2D(256, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(1e-4)))
model_5.add(BatchNormalization())
model_5.add(MaxPooling2D(2,2))

model_5.add(GlobalAveragePooling2D())
model_5.add(Dense(256, activation='relu'))
model_5.add(Dropout(0.5))
model_5.add(Dense(1, activation='sigmoid'))

model_5.compile(
    loss='binary_crossentropy',
    optimizer=RMSprop(learning_rate=1e-4),
    metrics=['accuracy']
)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, Dense, Dropout,
    BatchNormalization, GlobalAveragePooling2D
)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import RMSprop

# =========================
# MODEL 6 – bardzo głęboki CNN + silna regularyzacja
# =========================

model_6 = Sequential()

# Pierwsza warstwa (jedyna z input_shape)
model_6.add(Conv2D(32, (3,3), activation='relu', padding='same',
                   kernel_regularizer=l2(5e-4),
                   input_shape=(150,150,3)))
model_6.add(BatchNormalization())
model_6.add(MaxPooling2D(2,2))
model_6.add(Dropout(0.3))

# Kolejne bloki konwolucyjne
for filters in [64, 128, 256, 256]:
    model_6.add(Conv2D(filters, (3,3), activation='relu', padding='same',
                       kernel_regularizer=l2(5e-4)))
    model_6.add(BatchNormalization())
    model_6.add(MaxPooling2D(2,2))
    model_6.add(Dropout(0.3))

# Część klasyfikująca
model_6.add(GlobalAveragePooling2D())
model_6.add(Dense(128, activation='relu'))
model_6.add(Dropout(0.6))
model_6.add(Dense(1, activation='sigmoid'))

# Kompilacja (TYLKO RAZ)
model_6.compile(
    loss='binary_crossentropy',
    optimizer=RMSprop(learning_rate=5e-5),
    metrics=['accuracy']
)

# Podsumowanie modelu
model_6.summary()



```
