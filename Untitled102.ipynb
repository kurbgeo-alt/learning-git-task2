```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =====================================================
# 1. Wczytanie i czyszczenie danych
# =====================================================

data = pd.read_csv("HRDataset (2).csv")

# Usunięcie duplikatów kolumn
data = data.loc[:, ~data.columns.duplicated()]
data = data.reset_index(drop=True)

# Usunięcie białych znaków w kolumnach tekstowych
for col in data.select_dtypes(include="object").columns:
    data[col] = data[col].astype(str).str.strip()

# =====================================================
# 2. Przygotowanie zmiennych (Age, TenureYears)
# =====================================================

today = pd.Timestamp.today()

# Konwersje dat
data["DOB"] = pd.to_datetime(data["DOB"], errors="coerce")
data["DateofHire"] = pd.to_datetime(data["DateofHire"], errors="coerce")
data["DateofTerm"] = pd.to_datetime(data["DateofTerm"], errors="coerce")

# Wiek
data["Age"] = (today - data["DOB"]).dt.days // 365

# Staż pracy (aktywni → do dziś, zwolnieni → do DateofTerm)
data["TenureYears"] = (
    (data["DateofTerm"].fillna(today) - data["DateofHire"])
    .dt.days / 365.25
)

# Tylko aktywni pracownicy (do analiz bieżących)
active = data[data["EmploymentStatus"] == "Active"].copy()

# =====================================================
# 3. Wydajność vs Menedżer (kategorie opisowe)
# =====================================================

plt.figure(figsize=(12,6))
sns.countplot(
    x="ManagerName",
    hue="PerformanceScore",
    data=data
)
plt.xticks(rotation=90)
plt.title("Rozkład ocen wydajności w zależności od menedżera")
plt.tight_layout()
plt.show()

# =====================================================
# 4. Średni staż vs źródło rekrutacji
# =====================================================

mean_tenure = (
    data.groupby("RecruitmentSource", as_index=False)["TenureYears"]
    .mean()
    .sort_values("TenureYears", ascending=False)
)

plt.figure(figsize=(12,5))
sns.barplot(
    data=mean_tenure,
    x="RecruitmentSource",
    y="TenureYears"
)
plt.xticks(rotation=45)
plt.title("Średni staż pracowników w zależności od źródła rekrutacji")
plt.ylabel("Średni staż (lata)")
plt.tight_layout()
plt.show()

# =====================================================
# 5. Zadowolenie vs stan cywilny
# =====================================================

plt.figure(figsize=(10,6))
sns.boxplot(
    x="MaritalDesc",
    y="EmpSatisfaction",
    data=data
)
plt.title("Zadowolenie z pracy w zależności od stanu cywilnego")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# =====================================================
# 6. Wiek aktywnych pracowników
# =====================================================

plt.figure(figsize=(10,6))
plt.hist(
    active["Age"].dropna(),
    bins=12,
    color="skyblue",
    edgecolor="black",
    alpha=0.8
)
plt.title("Struktura wieku aktywnych pracowników")
plt.xlabel("Wiek")
plt.ylabel("Liczba pracowników")
plt.grid()
plt.tight_layout()
plt.show()

# =====================================================
# 7. Wiek vs liczba projektów specjalnych
# =====================================================

data["SpecialProjectsCount"] = pd.to_numeric(
    data["SpecialProjectsCount"],
    errors="coerce"
)

plt.figure(figsize=(10,6))
sns.regplot(
    x="Age",
    y="SpecialProjectsCount",
    data=data,
    scatter_kws={"alpha": 0.6}
)
plt.title("Zależność między wiekiem a liczbą projektów specjalnych")
plt.xlabel("Wiek")
plt.ylabel("Liczba projektów specjalnych")
plt.tight_layout()
plt.show()

# Korelacja
corr = data[["Age", "SpecialProjectsCount"]].corr().iloc[0,1]
print(f"Korelacja Pearsona (Age vs SpecialProjectsCount): {corr:.3f}")
```
