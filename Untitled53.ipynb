```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import requests
```

```python
from io import StringIO
```

```python
# 1. Wczytanie danych o interwencjach
# ==================================================

df = pd.read_csv("fatal-police-shootings-data (5).csv")

df["signs_of_mental_illness"] = (
    df["signs_of_mental_illness"]
    .astype(str)
    .str.lower() == "true"
)
```

```python
# 2. Zestawienie liczby ofiar wg rasy i chorob psychicznych
# ==================================================

table = (
    df.groupby(["race", "signs_of_mental_illness"])
    .size()
    .unstack(fill_value=0)
)

table["total"] = table.sum(axis=1)

table["percent_with_mental_illness"] = table.apply(
    lambda row: row[True] / row["total"] * 100,
    axis=1
)

print(table)
```

```python
# 3. Dodanie  kolumny dni tygodnia i zliczenie interwencji
# ==================================================

df["date"] = pd.to_datetime(df["date"])
df["day_of_week"] = df["date"].dt.day_name()

# ==================================================
ordered_days = [
    "Monday", "Tuesday", "Wednesday",
    "Thursday", "Friday", "Saturday", "Sunday"
]

counts_by_day = (
    df["day_of_week"]
    .value_counts()
    .reindex(ordered_days))
```

```python
# Wykres kolumnowy (poniedziaek  niedziela)
plt.figure(figsize=(10, 6))
counts_by_day.plot(kind="bar", color="steelblue")
plt.title("Liczba miertelnych interwencji policji wg dnia tygodnia")
plt.xlabel("Dzie tygodnia")
plt.ylabel("Liczba interwencji")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# 4. Wczytanie populacji stanow USA (Wikipedia)
# ==================================================

url_pop = "https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population"

html_pop = requests.get(
    url_pop,
    headers={"User-Agent": "Mozilla/5.0"}
).text

pop_tables = pd.read_html(StringIO(html_pop))
pop = pop_tables[0].copy()

pop.columns = ["_".join([str(x) for x in col if x]) for col in pop.columns]

state_col = [c for c in pop.columns if "state" in c.lower()][0]
pop_col = [c for c in pop.columns if "july" in c.lower()][0]

pop["population"] = (
    pop[pop_col]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.extract(r"(\d+)")
    .astype(float)
 )

pop["state"] = (
    pop[state_col]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.strip()
)

pop = pop[["state", "population"]]   
```

```python
# 5. Wczytanie skrotow stanow (Wikipedia)
# ==================================================

url_abbr = "https://en.wikipedia.org/wiki/List_of_U.S._state_and_territory_abbreviations"

html_abbr = requests.get(
    url_abbr,
    headers={"User-Agent": "Mozilla/5.0"}
).text

abbr_tables = pd.read_html(StringIO(html_abbr))

for t in abbr_tables:
    if any("USPS" in str(c) for c in t.columns):
        abbr = t.copy()
        break

if isinstance(abbr.columns, pd.MultiIndex):
    abbr.columns = [c[0] for c in abbr.columns]

state_col = [c for c in abbr.columns if "name" in c.lower()][0]
abbr_col = [c for c in abbr.columns if "usps" in c.lower()][0]

abbr = abbr.rename(columns={
    state_col: "state",
    abbr_col: "abbreviation"
})

abbr = abbr[["state", "abbreviation"]]

abbr["state"] = (
    abbr["state"]
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.strip()
)
```

```python
# 6. Incydenty na 1000 mieszkancow
# ==================================================

merged_pop = pop.merge(abbr, on="state", how="left")

counts_by_state = df["state"].value_counts().reset_index()
counts_by_state.columns = ["abbreviation", "incidents"]

final = counts_by_state.merge(
    merged_pop,
    on="abbreviation",
    how="left"
)

final = final[final["population"].notna()]

final["incidents_per_1000"] = (
    final["incidents"] / final["population"] * 1000
)

print(final.head())
```
