```python
import pandas as pd
```

```python
import plotly.graph_objects as go
```

```python
#Wczytanie danych
df_kghm =pd.read_csv('kgh_d (4).csv')
```

```python
df_copper = pd.read_csv('ca_c_f_d (6).csv')
```

```python
# WYKRES 1  Ceny zamknicia KGHM

fig1 = go.Figure()
fig1.add_trace(go.Scatter(
    x=df_kghm['Data'],
    y=df_kghm['Zamkniecie'],
    mode='lines',
    name='KGHM'
))

fig1.update_layout(
    title='Ceny zamknicia KGHM',
    xaxis_title='Rok',
    yaxis_title='Cena KGHM'
)

fig1.show()
```

```python
# WYKRES 2  Ceny miedzi
# =============================
fig2 = go.Figure()
fig2.add_trace(go.Scatter(
    x=df_copper['Data'],
    y=df_copper['Zamkniecie'],
    mode='lines',
    name='Mied',
    line=dict(color='orange')
))

fig2.update_layout(
    title='Ceny miedzi (zamknicie)',
    xaxis_title='Rok',
    yaxis_title='Cena miedzi'
)

fig2.show()
```

```python
#TABELA  poczenie danych KGHM i miedzi
# =============================
df_table = pd.merge(
    df_kghm[['Data', 'Zamkniecie']],
    df_copper[['Data', 'Zamkniecie']],
    on='Data',
    suffixes=('_KGHM', '_Miedz')
)

fig3 = go.Figure(
    data=[
        go.Table(
            header=dict(
                values=list(df_table.columns),
                fill_color='turquoise',
                align='left'
            ),
            cells=dict(
                values=[
                    df_table['Data'],
                    df_table['Zamkniecie_KGHM'],
                    df_table['Zamkniecie_Miedz']
                ],
                fill_color='orange',
                align='left'
            )
        )
    ]
)

fig3.update_layout(title='Tabela: Ceny zamknicia KGHM i ceny  miedzi')
fig3.show()
```
