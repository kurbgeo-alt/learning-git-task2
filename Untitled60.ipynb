```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.metr
ics import f1_score
```

```python
# Funkcja do wizualizacji granicy decyzyjnej

def plot_classification_surface(model, X, y, ax, title=""):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k")
    ax.set_title(title)
```

```python
# Generowanie danych

X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    class_sep=1.2,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Konfiguracja eksperymentu

Cs = [0.01, 0.1, 1, 10, 100]

configs = [
    {"name": "L2", "penalty": "l2", "solver": "lbfgs"},
    {"name": "L1", "penalty": "l1", "solver": "liblinear"},
    {"name": "ElasticNet", "penalty": "elasticnet", "solver": "saga", "l1_ratio": 0.5}
]

results = []
```

```python
# Trenowanie modeli

for cfg in configs:
    for C in Cs:

        clf_params = {
            "C": C,
            "penalty": cfg["penalty"],
            "solver": cfg["solver"],
            "max_iter": 5000
        }

        if cfg["penalty"] == "elasticnet":
            clf_params["l1_ratio"] = cfg["l1_ratio"]

        model = Pipeline([
            ("scaler", StandardScaler()),
            ("clf", LogisticRegression(**clf_params))
        ])

        model.fit(X_train, y_train)

        f1_train = f1_score(y_train, model.predict(X_train))
        f1_test = f1_score(y_test, model.predict(X_test))

        results.append({
            "Regularyzacja": cfg["name"],
            "C": C,
            "F1_train": f1_train,
            "F1_test": f1_test
        })


df = pd.DataFrame(results)
```

```python
# Wykres F1-score vs C

fig, ax = plt.subplots(figsize=(8, 5))

for reg in df["Regularyzacja"].unique():
    subset = df[df["Regularyzacja"] == reg]
    ax.plot(subset["C"], subset["F1_test"], marker="o", label=reg)

ax.set_xscale("log")
ax.set_xlabel("C (odwrotno siy regularyzacji)")
ax.set_ylabel("F1-score (test)")
ax.set_title("Wpyw regularyzacji na jako modelu")
ax.legend()
ax.grid(True)

plt.show()
```

```python
# Wizualizacja granic decyzyjnych najlepszych modeli

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

best_settings = [
    ("L2", "l2", "lbfgs", 1),
    ("L1", "l1", "liblinear", 0.1),
    ("ElasticNet", "elasticnet", "saga", 1)
]

for ax, (name, penalty, solver, C) in zip(axes, best_settings):

    params = {
        "C": C,
        "penalty": penalty,
        "solver": solver,
        "max_iter": 5000
    }

    if penalty == "elasticnet":
        params["l1_ratio"] = 0.5

    model = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(**params))
    ])

    model.fit(X_train, y_train)

    f1 = f1_score(y_test, model.predict(X_test))

    plot_classification_surface(
        model,
        X_test,
        y_test,
        ax,
        title=f"{name}\nC={C}, F1={f1:.3f}"
     )

plt.show()
```
