```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("HRDataset (2).csv", engine="python")

print("Kolumny w zbiorze:")
print(data.columns.tolist())data['DOB'] = pd.to_datetime(data['DOB'], errors='coerce')
data['DateofHire'] = pd.to_datetime(data['DateofHire'], errors='coerce')
data['DateofTermination'] = pd.to_datetime(data['DateofTermination'], errors='coerce')today = pd.to_datetime("2019-12-31")

data['Age'] = (today - data['DOB']).dt.days / 365
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')data['EndDate'] = data['DateofTermination'].fillna(today)
data['Tenure'] = (data['EndDate'] - data['DateofHire']).dt.days / 365data['SpecialProjectsCount'] = pd.to_numeric(data['SpecialProjectsCount'], errors='coerce')

corr = data['Age'].corr(data['SpecialProjectsCount'])
print("Korelacja (r) między wiekiem a liczbą projektów:", corr)

plt.figure()
plt.scatter(data['Age'], data['SpecialProjectsCount'])
plt.title(f"Wiek vs Liczba projektów specjalnych (r = {corr:.2f})")
plt.xlabel("Wiek")
plt.ylabel("Liczba projektów specjalnych")
plt.tight_layout()
plt.show()print("Unikalne wartości PerformanceScore:")
print(data['PerformanceScore'].unique())

performance_map = {
    'PIP': 1,
    'Needs Improvement': 2,
    'Fully Meets': 3,
    'Exceeds': 4
}

data['PerformanceScore_num'] = data['PerformanceScore'].map(performance_map)manager_perf = data.groupby('ManagerName')['PerformanceScore_num'].mean().sort_values()

plt.figure()
manager_perf.plot(kind='bar')
plt.title("Średnia PerformanceScore wg Managera")
plt.xlabel("Manager")
plt.ylabel("Średnia PerformanceScore")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()recruit_tenure = data.groupby('RecruitmentSource')['Tenure'].mean().sort_values(ascending=False)

plt.figure()
recruit_tenure.plot(kind='bar')
plt.title("Średni staż pracy wg źródła rekrutacji")
plt.xlabel("Recruitment Source")
plt.ylabel("Średni staż (lata)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()active = data[data['EmploymentStatus'] == 'Active']

plt.figure()
plt.hist(active['Age'], bins=8)
plt.title("Struktura wieku aktywnych pracowników")
plt.xlabel("Wiek")
plt.ylabel("Liczba pracowników")
plt.tight_layout()
plt.show()Wnioski

Korelacja między wiekiem a liczbą projektów specjalnych jest słaba / umiarkowana / silna (w zależności od wyniku), co oznacza że wiek ma / nie ma istotnego wpływu na liczbę powierzanych projektów.

Widoczne są różnice w średnich ocenach PerformanceScore między managerami, co może sugerować różnice w stylu zarządzania lub sposobie oceniania pracowników.

Źródło rekrutacji wpływa na długość zatrudnienia – pracownicy z wybranych źródeł wykazują dłuższy średni staż pracy.

Stan cywilny nie wykazuje silnego wpływu na poziom satysfakcji pracowników (różnice są niewielkie).

Struktura wieku aktywnych pracowników wskazuje na dominację określonej grupy wiekowej.
```
