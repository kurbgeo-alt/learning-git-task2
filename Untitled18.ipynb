```python
import pandas as pd
import matplotlib.pyplot as plt


movies = pd.read_csv("tmbd_movies(1).csv")
genres = pd.read_csv("tmbd_genres(1).csv")

#łączenie tabel
df = movies.merge(genres, on="genre_id", how="left")

df.head()
genre_counts = df["genre_name"].value_counts()
most_common_genre = genre_counts.index[0]
count_common = genre_counts.iloc[0]

print("Najczęściej występujący gatunek:", most_common_genre)
print("Liczba filmów tego gatunku:", count_common)
runtime_by_genre = df.groupby("genre_name")["runtime"].mean().sort_values(ascending=False)

longest_genre = runtime_by_genre.index[0]
longest_runtime = runtime_by_genre.iloc[0]

print("Gatunek o największym średnim czasie trwania:", longest_genre)
print("Średni czas:", longest_runtime)
subset = df[df["genre_name"] == longest_genre]

plt.figure(figsize=(10,6))
plt.hist(subset["runtime"].dropna(), bins=20, color="orange", edgecolor="black")
plt.title(f"Histogram czasu trwania filmów – gatunek: {longest_genre}")
plt.xlabel("Czas trwania (minuty)")
plt.ylabel("Liczba filmów")
plt.show()
```
