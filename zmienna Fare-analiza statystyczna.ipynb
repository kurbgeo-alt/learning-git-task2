```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


df = pd.read_csv('titanic.csv')


fare_values = df['Fare'].dropna()

mean_fare = np.mean(fare_values)
median_fare = np.median(fare_values)
mode_fare = stats.mode(fare_values, keepdims=True).mode[0]

q0 = np.min(fare_values)
q1 = np.quantile(fare_values, 0.25)
q3 = np.quantile(fare_values, 0.75)
q4 = np.max(fare_values)

print(f"Średnia: {mean_fare}")
print(f"Mediana: {median_fare}")
print(f"Moda: {mode_fare}")
print(f"Q0 (min): {q0}")
print(f"Q1: {q1}")
print(f"Q2 (mediana): {median_fare}")
print(f"Q3: {q3}")
print(f"Q4 (max): {q4}")


plt.figure(figsize=(10,5))
plt.hist(fare_values, bins=30, edgecolor='black', color='skyblue')
plt.axvline(mean_fare, color='red', label='Średnia')
plt.axvline(median_fare, color='green', label='Mediana')
plt.axvline(mode_fare, color='orange', label='Moda')
plt.legend()
plt.title("Histogram opłat (Fare)")
plt.xlabel("Fare")
plt.ylabel("Liczba pasażerów")
plt.show()

# Wykres pudełkowy
plt.figure(figsize=(8,4))
sns.boxplot(x=fare_values, color='lightgreen')
plt.title("Wykres pudełkowy zmiennej Fare")
plt.xlabel("Fare")
plt.show()
```
